## I. Generate an n-dimensional random vector ùíó = [ùë£_1, ùë£_2,‚Ä¶, ùë£_n)] with non-negative elements. For ùíó, implement the following calculations and algorithms:

```python
import random
import timeit
import numpy as np
import matplotlib.pyplot as plt
```

```python
def generate_random_vector(n):
    return [random.randint(0, 100) for _ in range(n)]

random_vectors = []
for n in range(1, 2001):
    v = generate_random_vector(n)
    random_vectors.append(v)
```

```python
def constant_function(v):
    return 2  # Replace 42 with your desired constant value

execution_times_constant = []
for v in random_vectors:
    times = []
    for _ in range(5):
        start_time = timeit.default_timer()
        constant_function(v)
        end_time = timeit.default_timer()
        times.append(end_time - start_time)
    execution_times_constant.append(np.mean(times))

# Plotting the results
plt.plot(range(1, 2001), execution_times_constant)
plt.xlabel('Dimension (n)')
plt.ylabel('Execution Time (seconds)')
plt.title('Constant Function Execution Time')
plt.show()
```

```python
def summation_function(v):
    return sum(v)

execution_times_summation = []
for v in random_vectors:
    times = []
    for _ in range(5):
        start_time = timeit.default_timer()
        summation_function(v)
        end_time = timeit.default_timer()
        times.append(end_time - start_time)
    execution_times_summation.append(np.mean(times))

# Plotting the results
plt.plot(range(1, 2001), execution_times_summation)
plt.xlabel('Dimension (n)')
plt.ylabel('Execution Time (seconds)')
plt.title('Summation Function Execution Time')
plt.show()
```

```python
def product_function(v):
    product = 1
    for element in v:
        product *= element
    return product

execution_times_product = []
for v in random_vectors:
    times = []
    for _ in range(5):
        start_time = timeit.default_timer()
        product_function(v)
        end_time = timeit.default_timer()
        times.append(end_time - start_time)
    execution_times_product.append(np.mean(times))

# Plotting the results
plt.plot(range(1, 2001), execution_times_product)
plt.xlabel('Dimension (n)')
plt.ylabel('Execution Time (seconds)')
plt.title('Product Function Execution Time')
plt.show()
```

```python
def polynomial_direct_calculation(v, x):
    degree = len(v) - 1
    result = 0
    for i in range(degree, -1, -1):
        result = result * x + v[i]
    return result

evaluation_times_direct = []
for v in random_vectors:
    times = []
    for _ in range(5):
        start_time = timeit.default_timer()
        polynomial_direct_calculation(v, 1.5)
        end_time = timeit.default_timer()
        times.append(end_time - start_time)
    evaluation_times_direct.append(np.mean(times))

# Plotting the results
plt.plot(range(1, 2001), evaluation_times_direct)
plt.xlabel('Dimension (n)')
plt.ylabel('Execution Time (seconds)')
plt.title('Polynomial Evaluation (Direct Calculation) Time Complexity')
plt.show()
```

```python
def polynomial_horner_method(v, x):
    result = 0
    for coeff in reversed(v):
        result = result * x + coeff
    return result

evaluation_times_horner = []
for v in random_vectors:
    times = []
    for _ in range(5):
        start_time = timeit.default_timer()
        polynomial_horner_method(v, 1.5)
        end_time = timeit.default_timer()
        times.append(end_time - start_time)
    evaluation_times_horner.append(np.mean(times))

# Plotting the results
plt.plot(range(1, 2001), evaluation_times_horner)
plt.xlabel('Dimension (n)')
plt.ylabel('Execution Time (seconds)')
plt.title("Polynomial Evaluation (Horner's Method) Time Complexity")
plt.show()
```

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]

def analyze_bubble_sort():
    n_values = range(1, 501)
    execution_times = []
    runs = 5

    for n in n_values:
        total_time = 0
        for _ in range(runs):
            random_vector = np.random.rand(n)
            start_time = time.time()
            bubble_sort(random_vector)
            end_time = time.time()
            elapsed_time = end_time - start_time
            total_time += elapsed_time
        average_time = total_time / runs
        execution_times.append(average_time)

    return execution_times

execution_times = analyze_bubble_sort()
n_values = range(1, 501)

plt.plot(n_values, execution_times)
plt.xlabel('Size of Vector (n)')
plt.ylabel('Execution Time (seconds)')
plt.title('Bubble Sort Execution Time Analysis')
plt.show()
```

```python
import numpy as np
import time
import matplotlib.pyplot as plt

def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quick_sort(left) + middle + quick_sort(right)

def analyze_quick_sort():
    n_values = range(1, 2001)
    execution_times = []
    runs = 5

    for n in n_values:
        total_time = 0
        for _ in range(runs):
            random_vector = np.random.rand(n)
            start_time = time.time()
            quick_sort(random_vector)
            end_time = time.time()
            elapsed_time = end_time - start_time
            total_time += elapsed_time
        average_time = total_time / runs
        execution_times.append(average_time)

    return execution_times

execution_times = analyze_quick_sort()
n_values = range(1, 2001)

plt.plot(n_values, execution_times)
plt.xlabel('Size of Vector (n)')
plt.ylabel('Execution Time (seconds)')
plt.title('Quick Sort Execution Time Analysis')
plt.show()
```

## II. Generate random matrices ùê¥ and ùêµ of size ùëõ √ó ùëõ with non-negative elements. Find the usual matrix product for ùê¥ and ùêµ.

```python
# Function to generate a random matrix with a given number of non-zero elements
def generate_random_matrix(n, num_nonzero):
    matrix = np.zeros((n, n))
    indices = random.sample(range(n*n), num_nonzero)
    for index in indices:
        row = index // n
        col = index % n
        matrix[row][col] = random.randint(1, 10)
    return matrix

# Matrix multiplication
def matrix_multiplication(matrix1, matrix2):
    return np.dot(matrix1, matrix2)

# Perform time complexity analysis for Cartesian product of matrices
def analyze_cartesian_product_time():
    matrix_sizes = list(range(1, 2001))
    execution_times = []

    for size in matrix_sizes:
        matrix1 = generate_random_matrix(size, size)
        matrix2 = generate_random_matrix(size, size)

        start_time = time.time()
        matrix_multiplication(matrix1, matrix2)
        end_time = time.time()
        execution_time = end_time - start_time

        execution_times.append(execution_time)

    return matrix_sizes, execution_times

# Plot the time complexity analysis results
def plot_time_complexity_analysis(matrix_sizes, execution_times):
    plt.plot(matrix_sizes, execution_times)
    plt.xlabel("Matrix Size")
    plt.ylabel("Execution Time (seconds)")
    plt.title("Time Complexity Analysis of Cartesian Product")
    plt.show()

# Perform time complexity analysis and plot the results
matrix_sizes, execution_times = analyze_cartesian_product_time()
plot_time_complexity_analysis(matrix_sizes, execution_times)
```
